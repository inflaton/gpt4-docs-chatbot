## Testing with Chrome Automation

Please refer to [official Selenium Python doc](https://selenium-python.readthedocs.io/installation.html) to install & run Chrome driver

1. First import the WebDriver and Keys classes from Selenium.

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

2. Next, create an instance of Chrome with the path of the driver that you downloaded through the websites of the respective browser. In this example, we assume that the driver is in the same directory as the Python script that you will execute.

```python
driver = webdriver.Chrome()
```

3. Set up

```python
driver.get("http://localhost:3000/")
```

4. Define functions for test case

```python
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.by import By

def query(question):
    wait = WebDriverWait(driver, 160)
    user_input = wait.until(EC.element_to_be_clickable((By.ID, 'userInput')))

    print("asking question: " + question)

    user_input = driver.find_element("name", "userInput")
    user_input.clear()
    user_input.send_keys(question)
    user_input.send_keys(Keys.RETURN)

def run(vectorStore):
    print("run test case with vector store: " + vectorStore)
    driver.get("http://localhost:3000/")

    query("What is Mastercard Priceless?")
    query("Dining in New York")
    query("Russian Tea Room?")
    query("请推荐一下纽约的娱乐")

    wait = WebDriverWait(driver, 160)
    user_input = wait.until(EC.element_to_be_clickable((By.ID, 'userInput')))
```

5. Run test cases

Before running the test case, please update .env file to make sure `VECTOR_STORE=hnswlib` is set properly and then run `yarn hnswlib` which will dump logs to file data/logs/hnswlib.log.

```python
%%time

run("hnswlib")
```

Before running the test case, please update .env file to make sure `VECTOR_STORE=chroma` is set properly and then run `yarn chroma` which will dump logs to file data/logs/chroma.log.

```python
%%time

run("chroma")
```

Before running the test case, please update .env file to make sure `VECTOR_STORE=milvus` is set properly and then run `yarn milvus` which will dump logs to file data/logs/milvus.log.

```python
%%time

run("milvus")
```

Before running the test case, please update .env file to make sure `VECTOR_STORE=pinecone` is set properly and then run `yarn pinecone` which will dump logs to file data/logs/pinecone.log.

```python
%%time

run("pinecone")
```

5. Tear down

```python
driver.close()
```
